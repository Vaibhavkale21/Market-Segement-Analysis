```python
def get_state_markets(row):
    if row['market'].strip() in kerala_market:
        val = 'Kerala Market'
    elif row['market'].strip()  in AP_market:
        val = 'Andhra Pradesh Market'
    elif row['market'].strip()  in TN_market:
        val = 'Tamil Nadu Market'
    elif row['market'].strip()  in UP_market:
        val = 'Uttar Pradesh Market'
    elif row['market'].strip()  in Assam_market:
        val = 'Assam_market'
    
    else:
        val = 'No Value'
    return val

drop_df['State Markets'] = drop_df.apply(get_state_markets, axis=1)
drop_df['State Markets'].unique()
drop_df
```

```python
def get_commodity_categories(row):
    if row['commodity'].strip() in fruits:
        val = 'FRUIT'
    elif row['commodity'].strip()  in vegetables:
        val = 'VEGETABLE'
    elif row['commodity'].strip()  in cereals:
        val = 'CEREAL'
    elif row['commodity'].strip()  in pulses:
        val = 'PULSE'
    elif row['commodity'].strip()  in spices:
        val = 'SPICE'
    elif row['commodity'].strip()  in millets:
        val = 'MILLET'
    else:
        val = 'OTHER'
    return val

drop_df['Com_Category'] = drop_df.apply(get_commodity_categories, axis=1)
drop_df['Com_Category'].unique()
drop_df
```

```python
col_with_str=final_df[['state','district','market','commodity']]
```

```python
dum_var=pd.get_dummies(col_with_str)
dum_var
```

```python
dropped_cat=final_df.drop(col_with_str,1)
temp=dropped_cat
final_dataset=pd.concat([dum_var,temp], axis=1)
final_dataset
```

```python
from sklearn.model_selection import train_test_split
x=final_dataset.drop('modal_price',1)
y=final_dataset['modal_price']
```

```python
x.shape
```

```python
y.shape
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(train_x, train_y)
print(regr.score(test_x, test_y))
print(regr.coef_) 
print('Intercept: \n', regr.intercept_)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_jobs=-1,
                              random_state=42)
model.fit(train_x,train_y)
```

```python
model.score(test_x,test_y)
```

```python
from sklearn.linear_model import LogisticRegression
LogR=LogisticRegression(max_iter=1000,random_state=42,n_jobs=-1)
LogR.fit(train_x, train_y)
print(LogR.score(test_x, test_y))
```
